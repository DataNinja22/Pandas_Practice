```python
# Importing required libraries for data manipulation and analysis
# pandas: Used for data manipulation and analysis with DataFrames
# numpy: Used for numerical computing and array operations
import pandas as pd
import numpy as np
```

```python
# Loading the movies dataset from CSV file using pandas read_csv() function
# read_csv(): Reads a comma-separated values (CSV) file into a DataFrame
# The dataset contains IMDb top 1000 movies with various attributes like Genre, Rating, etc.
movies = pd.read_csv('DataSets/dataset-session-19/imdb-top-1000.csv')
```

## Remember group by only on the categorical things

```python
# Display the first few rows of the movies dataset using head() function
# head(): Returns the first n rows of DataFrame (default n=5)
# This helps in understanding the structure and content of the dataset
movies.head()
```

```python
# Create a GroupBy object by grouping movies by the 'Genre' column
# groupby(): Groups DataFrame using a mapper or by a Series of columns
# This creates a GroupBy object that allows for aggregation operations on each genre
# The groupby operation is lazy - it doesn't compute results until an aggregation is applied
genres = movies.groupby('Genre')  # Grouping by 'Genre' column
```

```python
# Apply sum aggregation on the GroupBy object to get totals for each genre
# sum(): Computes the sum of numeric columns for each group
# This will sum all numeric columns (Runtime, IMDB_Rating, No_of_Votes, Gross, etc.) by genre
# Non-numeric columns are ignored in the aggregation
genres.sum()
```

```python
# Get the minimum value in each group using min aggregation
# min(): Returns the minimum value for each numeric column within each group
# This finds the lowest values (e.g., shortest runtime, lowest rating) for each genre
genres.min()  # Minimum value in each group
```

```python
# Find the top 3 genres by total earnings using chained operations
# groupby(): Groups movies by Genre
# sum(): Calculates sum of all numeric columns for each genre
# ['Gross']: Selects only the Gross earnings column from the aggregated result
# sort_values(): Sorts values in descending order (highest earnings first)
# head(3): Returns the top 3 results
movies.groupby('Genre').sum()['Gross'].sort_values(ascending=False).head(3)
```

```python
# Find the top 3 genres by total earnings with direct column selection (more efficient)
# groupby(): Groups movies by Genre
# ['Gross']: Selects only the Gross column before aggregation (more efficient than selecting after)
# sum(): Calculates sum of Gross earnings for each genre
# sort_values(): Sorts in descending order
# head(3): Returns top 3 results
movies.groupby('Genre')['Gross'].sum().sort_values(ascending=False).head(3)
```

```python
# Find the genre with the highest average IMDB rating
# groupby(): Groups movies by Genre
# ['IMDB_Rating']: Selects only the IMDB_Rating column
# mean(): Calculates the average rating for each genre
# sort_values(): Sorts in descending order (highest average first)
# head(1): Returns only the top result (genre with highest average rating)
movies.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
```

```python
# Find the director with the highest total number of votes across all their movies
# groupby(): Groups movies by Director
# ['No_of_Votes']: Selects only the No_of_Votes column
# sum(): Calculates total votes received by each director across all their movies
# sort_values(): Sorts in descending order (highest total votes first)
# head(1): Returns only the director with the most total votes
movies.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head(1)
```

```python
# Example of finding the highest rated movie of each genre (commented out for reference)
# groupby(): Groups movies by Genre
# ['IMDB_Rating']: Selects only the IMDB_Rating column
# max(): Would return the maximum rating for each genre
# This shows the highest rating achieved in each genre, but not which movie achieved it
# movies.groupby('Genre')['IMDB_Rating'].max()
```

```python
# Count the number of movies per actor using two different approaches:

# Approach 1 (commented): Using value_counts() on Star1 column
# value_counts(): Returns a Series containing counts of unique values
# movies['Star1'].value_counts()

# Approach 2: Using groupby and count aggregation
# groupby(): Groups movies by Star1 (lead actor)
# ['Series_Title']: Selects the Series_Title column to count
# count(): Counts non-null values in each group (number of movies per actor)
# sort_values(): Sorts in descending order (actors with most movies first)
movies.groupby('Star1')['Series_Title'].count().sort_values(ascending=False)
```

```python
# GroupBy Attributes and Methods demonstration:
# This cell serves as documentation for various GroupBy operations available in pandas:

# Key GroupBy Attributes and Methods:
# - len(groupby_obj): Returns the total number of groups created
# - .size(): Returns the count of items in each group (includes NaN values)
# - .count(): Returns the count of non-null values in each group
# - .first(): Returns the first row from each group
# - .last(): Returns the last row from each group  
# - .nth(n): Returns the nth row from each group (0-based indexing)
# - .get_group(key): Returns the group corresponding to a specific key
# - .groups: Returns a dictionary with group labels as keys and row indices as values
# - .describe(): Returns descriptive statistics for each group
# - .sample(n): Randomly samples n rows from each group
# - .nunique(): Returns the number of unique values in each group
# - .agg() or .aggregate(): Applies one or more aggregation functions to groups

# Additional methods: sum(), mean(), median(), std(), var(), min(), max(), etc.
```

```python
# Display the total number of groups by Genre
# len(): Returns the count of groups in the GroupBy object
# This tells us how many unique genres exist in the dataset
# Each unique genre becomes one group in the GroupBy object
len(movies.groupby('Genre'))  # Total number of groups
```

```python
# Display the number of unique genres using Series method
# nunique(): Returns the number of unique non-null values in a Series
# This is an alternative way to count unique genres without using groupby
# Should return the same result as len(movies.groupby('Genre'))
movies['Genre'].nunique()
```

```python
# Display the size (number of movies) of each genre group
# size(): Returns the size of each group (counts all rows including NaN values)
# This shows how many movies belong to each genre
# Unlike count(), size() includes rows with missing values
movies.groupby('Genre').size() # Number of items in each group
```

```python
# Demonstrate methods to access specific rows in each group:
# Using nth(6) to get the 7th row (0-based index) of each genre group.
genres = movies.groupby('Genre')

# Available positional access methods:
# first(): Returns the first row of each group
# last(): Returns the last row of each group
# nth(n): Returns the nth row of each group (0-based indexing)

# genres.first()  # Uncomment to see first row of each genre
# genres.last()   # Uncomment to see last row of each genre

# Using nth(6) to get the 7th row (0-based index) of each genre group
# If a group has fewer than 7 rows, it will return NaN for that group
genres.nth(6)
```

```python
# Display count of movies for each genre using value_counts()
# value_counts(): Returns a Series with counts of unique values in descending order
# This is equivalent to groupby('Genre').size().sort_values(ascending=False)
# but more concise and automatically sorted in descending order
# Shows which genres have the most movies in the dataset
movies['Genre'].value_counts()
```

```python
# Retrieve data for the 'Fantasy' genre using two equivalent approaches:

# Approach 1: Using get_group() method on the GroupBy object
# get_group(key): Returns the group corresponding to the specified key
# More efficient when you have an existing GroupBy object
genres.get_group('Fantasy')

# Approach 2: Using boolean indexing on the original DataFrame
# Creates a boolean mask and filters the DataFrame
# This approach doesn't require a pre-existing GroupBy object
movies[movies['Genre'] == 'Fantasy']
```

```python
# Display the groups as a dictionary with group labels as keys and row indices as values
# .groups: Returns a dictionary where:
#   - Keys are the group labels (e.g., genre names)
#   - Values are arrays of row indices belonging to each group
# This shows the internal structure of how pandas organizes the GroupBy object
# Useful for understanding which rows belong to which groups
genres.groups
```

```python
# Get comprehensive descriptive statistics for each genre group
# describe(): Returns count, mean, std, min, 25%, 50%, 75%, max for numeric columns
# Applied to GroupBy object, it provides these statistics for each group separately
# This gives a statistical summary of all numeric columns (Runtime, IMDB_Rating, etc.) by genre
# Helpful for understanding the distribution and characteristics of each genre
genres.describe()  # Descriptive statistics for each group
```

```python
# Randomly sample 2 rows from each group with replacement
# sample(n, replace=True): Randomly selects n rows from each group
# replace=True: Allows the same row to be selected multiple times (sampling with replacement)
# replace=False (default): Each row can only be selected once (sampling without replacement)
# This is useful for getting a random representative sample from each genre
genres.sample(2, replace=True) # Randomly sample 2 rows from each group, with replacement
```

```python
# Count the number of unique values in each column for each genre group
# nunique(): Returns the number of unique non-null values for each column within each group
# This shows diversity within each genre (e.g., how many unique directors, actors, etc.)
# Useful for understanding the variety of content within each genre category
genres.nunique() # Count unique values in each group
```

```python
# Aggregate multiple columns with different functions using a dictionary mapping
# agg(): Applies different aggregation functions to different columns
# Dictionary format: {'column_name': 'aggregation_function'}
# This allows customized analysis - different metrics for different attributes:
# - Runtime: average duration of movies in each genre
# - IMDB_Rating: average quality rating per genre  
# - No_of_Votes: total audience engagement per genre
# - Gross: total box office earnings per genre
# - Metascore: lowest critic score in each genre
genres.agg(  #  Aggregate multiple columns with different functions
    {
        'Runtime': 'mean',      # Average movie duration per genre
        'IMDB_Rating': 'mean',  # Average rating per genre
        'No_of_Votes': 'sum',   # Total votes per genre
        'Gross': 'sum',         # Total earnings per genre
        'Metascore': 'min'      # Lowest critic score per genre
    }
)
```

```python
# Apply multiple aggregation functions to all numeric columns simultaneously
# agg(): When passed a list of functions, applies each function to every numeric column
# This creates a multi-level column structure (function_name, column_name)
# Functions applied: min, max, mean, sum for each numeric column in each genre
# Results in comprehensive statistical summary across all genres and metrics
genres.agg(['min', 'max', 'mean', 'sum'])
```

```python
# Advanced aggregation: applying different combinations of functions to specific columns
# agg(): Dictionary with column names as keys and function lists/strings as values
# This approach combines the flexibility of both previous methods:
# - Some columns get multiple functions (Runtime: min and mean)
# - Some columns get single functions (IMDB_Rating: mean only)
# - No_of_Votes gets both sum (total engagement) and max (peak popularity)
# Creates a multi-level structure only for columns with multiple functions
genres.agg(
    {
        'Runtime': ['min', 'mean'],    # Shortest and average runtime per genre
        'IMDB_Rating': 'mean',         # Average rating per genre  
        'No_of_Votes': ['sum', 'max'], # Total and peak votes per genre
        'Gross': 'sum',                # Total earnings per genre
        'Metascore': 'min'             # Lowest critic score per genre
    }
)
```

```python
# Find the highest rated movie in each genre using iteration and concatenation
# This approach gets the actual movie records (not just the rating values)

# Step 1: Initialize empty list to store DataFrames
df_list = []

# Step 2: Iterate through each group in the GroupBy object
# for group, data: 'group' is the genre name, 'data' is the DataFrame for that genre
for group, data in genres:
  # Find rows where IMDB_Rating equals the maximum rating in this genre
  # data['IMDB_Rating'].max(): Gets the highest rating in current genre
  # data[condition]: Filters to rows matching the condition (highest rated movies)
  df_list.append(data[data['IMDB_Rating'] == data['IMDB_Rating'].max()])

# Step 3: Combine all DataFrames into a single result
# pd.concat(): Concatenates DataFrames vertically (replaces deprecated df.append())
df = pd.concat(df_list)
df
```

```python
# Apply the built-in min function to each group using apply()
# apply(): Applies a function to each group and returns the result
# min: Built-in Python function that finds the minimum value
# When applied to GroupBy, min() operates on each group's DataFrame
# This is equivalent to genres.min() but demonstrates apply() functionality
# apply() is more flexible and can work with custom functions
genres.apply(min)
```

```python
# Define a custom function that counts movies starting with 'A' in a group
def foo(group):
  # group: DataFrame containing all movies for a specific genre
  # str.startswith('A'): String method that returns boolean Series (True if starts with 'A')
  # sum(): Counts True values (True=1, False=0) to get total count
  # This function will count how many movie titles start with 'A' in each genre
  return group['Series_Title'].str.startswith('A').sum()
```

```python
# Apply the custom function 'foo' to each genre group
# apply(): Calls the foo function once for each genre group
# Each genre's DataFrame is passed as the 'group' parameter to foo()
# Returns a Series with genre names as index and count of 'A' movies as values
# This shows how many movies starting with 'A' exist in each genre
genres.apply(foo)
```

```python
# Define a function to rank movies within each group based on IMDB rating
def rank_movie(group):
  # group: DataFrame containing all movies for a specific genre
  # rank(): Assigns ranks to values (1 = highest rating, 2 = second highest, etc.)
  # ascending=False: Higher ratings get lower rank numbers (1 is best)
  # Creates a new column 'genre_rank' showing each movie's ranking within its genre
  group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
  return group
```

```python
# Apply the ranking function to each genre group
# apply(): Calls rank_movie function for each genre group
# Returns the complete DataFrame with a new 'genre_rank' column added
# Each movie now has a rank within its genre (1 = best rated in that genre)
# This allows comparison of movies within their genre categories
genres.apply(rank_movie)
```

```python
# Define a function to normalize IMDB ratings within each genre group
def normal(group):
  # Normalization formula: (value - min) / (max - min)
  # This scales all ratings in a genre to a 0-1 range:
  # - Lowest rated movie in genre gets 0
  # - Highest rated movie in genre gets 1  
  # - Other movies get proportional values between 0 and 1
  # This allows fair comparison across genres with different rating ranges
  group['norm_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min()) / (group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
  return group

# Apply normalization function to each genre group
# Creates 'norm_rating' column showing relative performance within each genre
genres.apply(normal)
```

```python
# Perform groupby on multiple columns: Director and Star1 (lead actor)
# groupby(): Can accept a list of columns to create hierarchical groups
# This creates groups for each unique combination of Director and Star1
# Useful for analyzing director-actor collaborations
duo = movies.groupby(['Director', 'Star1'])

# Display the GroupBy object information
duo

# size(): Shows the number of movies for each director-actor combination
# Identifies prolific collaborations (multiple movies together)
duo.size()

# get_group(): Access a specific director-actor combination
# Tuple format: (Director_name, Star1_name) - both must match exactly
# This retrieves all movies where Aamir Khan directed and Amole Gupte starred
duo.get_group(('Aamir Khan', 'Amole Gupte'))
```

```python
# Find the most profitable director-actor combination by total box office earnings
# duo: GroupBy object with director-actor combinations
# ['Gross']: Selects only the Gross earnings column
# sum(): Calculates total earnings for each director-actor pair across all their collaborations
# sort_values(): Sorts in descending order (highest earning partnerships first)
# head(1): Returns only the top earning director-actor combination
duo['Gross'].sum().sort_values(ascending=False).head(1)
```

```python
# Find the best actor-genre combination based on average Metascore (critic ratings)
# groupby(): Groups by both Star1 (actor) and Genre
# ['Metascore']: Selects only the Metascore column (critic ratings)
# mean(): Calculates average critic score for each actor-genre combination
# reset_index(): Converts the result back to a regular DataFrame (removes multi-index)
# sort_values(): Sorts by Metascore in descending order (highest critic scores first)
# head(1): Returns the actor-genre combination with the highest average critic rating
movies.groupby(['Star1', 'Genre'])['Metascore'].mean().reset_index().sort_values('Metascore', ascending=False).head(1)
```

```python
# Apply multiple aggregation functions to all numeric columns for director-actor pairs
# duo: GroupBy object with director-actor combinations  
# agg(): Applies multiple functions to each numeric column
# ['min', 'max', 'mean']: Three statistical functions applied simultaneously
# Results show the minimum, maximum, and average values for each metric
# This provides comprehensive statistics for each director-actor collaboration
# Useful for understanding the range and consistency of their work together
duo.agg(['min', 'max', 'mean'])
```

```python
# Load the IPL (Indian Premier League) deliveries dataset for cricket analysis
# read_csv(): Loads cricket ball-by-ball data from CSV file
# This dataset contains detailed information about each ball bowled in IPL matches
# Including batsman, bowler, runs scored, extras, dismissals, etc.
ipl = pd.read_csv('/content/deliveries.csv')

# head(): Display first few rows to understand the dataset structure
# Shows columns like match_id, inning, batting_team, bowling_team, over, ball, batsman, etc.
ipl.head()
```

```python
# Display the dimensions (shape) of the IPL dataset
# shape: Returns a tuple (rows, columns) showing dataset size
# Rows represent the total number of balls bowled in the dataset
# Columns represent the number of attributes/features recorded for each ball
# This helps understand the scale of data we're working with
ipl.shape
```

```python
# Find the top 10 batsmen based on total runs scored across all IPL matches
# groupby('batsman'): Groups all deliveries by batsman name
# ['batsman_runs']: Selects the column containing runs scored off each ball
# sum(): Calculates total runs scored by each batsman across their career
# sort_values(): Sorts in descending order (highest run-scorers first)
# head(10): Returns the top 10 highest run-scorers in IPL history
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# Identify the batsman with the highest count of sixes (6 runs in a single ball)

# Step 1: Filter dataset to include only deliveries where 6 runs were scored
# ipl['batsman_runs'] == 6: Creates boolean mask for six-scoring deliveries
six = ipl[ipl['batsman_runs'] == 6]

# Step 2: Count sixes for each batsman and find the leader
# groupby('batsman'): Groups six-scoring deliveries by batsman
# ['batsman'].count(): Counts the number of six-scoring deliveries for each batsman
# sort_values(): Sorts in descending order (most sixes first)
# head(1): Gets the top result (batsman with most sixes)
# .index[0]: Extracts just the batsman name from the result
six.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1).index[0]
```

```python
# Find the batsman with the most boundaries (fours and sixes) in death overs (last 5 overs)

# Step 1: Filter for death overs (overs 16-20 in T20 cricket)
# over > 15: Selects deliveries from over 16 onwards (death overs)
temp_df = ipl[ipl['over'] > 15]

# Step 2: Filter for boundaries only (fours and sixes)
# batsman_runs == 4: Four runs scored (boundary)
# batsman_runs == 6: Six runs scored (maximum)
# |: OR operator to include both fours and sixes
temp_df = temp_df[(temp_df['batsman_runs'] == 4) | (temp_df['batsman_runs'] == 6)]

# Step 3: Count boundaries per batsman in death overs
# groupby('batsman'): Groups by batsman name
# count(): Counts total boundaries (4s + 6s) for each batsman in death overs
# sort_values(): Sorts in descending order
# head(1).index[0]: Gets the name of batsman with most death-over boundaries
temp_df.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1).index[0]
```

```python
# Calculate Virat Kohli's total runs scored against each bowling team

# Step 1: Filter dataset for Virat Kohli's batting performances only
# ipl['batsman'] == 'V Kohli': Boolean mask to select only Kohli's deliveries
temp_df = ipl[ipl['batsman'] == 'V Kohli']

# Step 2: Group by bowling team and sum runs
# groupby('bowling_team'): Groups Kohli's deliveries by the team he was batting against
# ['batsman_runs']: Selects the runs column
# sum(): Calculates total runs scored against each bowling team
# reset_index(): Converts result to a DataFrame with bowling_team and batsman_runs columns
# This shows Kohli's performance (total runs) against each IPL team
temp_df.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
# Find Virat Kohli's highest individual score in any single IPL match

# Step 1: Filter for Virat Kohli's batting performances
temp_df = ipl[ipl['batsman'] == 'V Kohli']

# Step 2: Calculate total runs per match and find the highest
# groupby('match_id'): Groups Kohli's deliveries by individual matches
# ['batsman_runs'].sum(): Calculates total runs scored in each match
# sort_values(): Sorts match scores in descending order (highest first)
# head(1): Gets the highest scoring match
# .values[0]: Extracts the actual score value (not the Series)
# This gives Kohli's highest individual match score in IPL
temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
# Define a reusable function to get the highest match score for any given batsman
def highest(batsman):
  # batsman: Parameter to specify which batsman's highest score to find
  
  # Step 1: Filter dataset for the specified batsman's performances
  temp_df = ipl[ipl['batsman'] == batsman]
  
  # Step 2: Calculate and return highest individual match score
  # groupby('match_id'): Groups by individual matches for this batsman
  # ['batsman_runs'].sum(): Calculates total runs in each match
  # sort_values(): Sorts match scores in descending order
  # head(1).values[0]: Returns the highest score value
  # This function can be called for any batsman to find their best performance
  return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
# Display the highest score for 'DA Warner'
# highest(): Calls the previously defined function
# 'DA Warner': Player name parameter (must match exactly as stored in dataset)
# This will return Warner's best individual match performance in terms of runs scored
# The function encapsulates the complex groupby logic for easy reuse with different players
highest('DA Warner')
```
